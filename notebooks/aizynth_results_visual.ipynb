```python
import json
from aizynthfinder.reactiontree import ReactionTree
from IPython.display import display

with open("outputs/26-06-12-hesperidin_combined_curated.json") as f:
    routes = json.load(f)

for i, route_dict in enumerate(routes):
    rt = ReactionTree.from_dict(route_dict)
    print(f"Route {i+1}: solved={rt.is_solved}, steps={len(list(rt.reactions()))}")
    display(rt.to_image())
```
