导入橙皮苷分子

```python
from rdkit import Chem
from rdkit.Chem import AllChem

hesperidin = Chem.MolFromMolFile('Hesperidin.sdf')
Chem.MolToSmiles(hesperidin)
```

导入橙皮素分子

```python
hesperitin = Chem.MolFromMolFile('Hesperitin.sdf')
Chem.MolToSmiles(hesperitin)
```

绘图橙皮苷分子和橙皮素分子

```python
from rdkit.Chem import Draw
Draw.MolToFile(hesperidin, 'hesperidin.png', (600, 400), kekulize=True)
Draw.MolToFile(hesperitin, 'hesperitin.png', (600, 400), kekulize=True)
```

绘图带橙皮素分子的RDKit读取序号的Smiles

```python
# MolToSmiles 默认返回 Canonical SMILES，原子顺序会被重排，但结构等价。
# 加上 isomericSmiles=True 可保留 @ 手性信息。
print(Chem.MolToSmiles(hesperitin, isomericSmiles=True))
```

```python
from rdkit.Chem.Draw import rdMolDraw2D
 
#  绘图方式：在原子旁边标注 1-based 序号
mol_copy = Chem.Mol(hesperitin)
for atom in mol_copy.GetAtoms():
    atom.SetProp('atomNote', str(atom.GetIdx() + 1))

drawer = rdMolDraw2D.MolDraw2DCairo(600, 600)
drawer.drawOptions().annotationFontScale = 0.75
rdMolDraw2D.PrepareAndDrawMolecule(drawer, mol_copy)
drawer.WriteDrawingText('hesperitin_atom_1based.png')
```

带Smiles原子序号的结构

```python
import re
from rdkit import Chem
from rdkit.Chem.Draw import rdMolDraw2D

smiles = "COc1ccc([C@@H]2CC(=O)c3c(O)cc(O)cc3O2)cc1O"

# 1. 从原始 SMILES 直接构建 mol
#    此时 atom.GetIdx() 就是 SMILES 中出现顺序的 0-based 索引
mol = Chem.MolFromSmiles(smiles)

# 2. （可选）用正则提取 SMILES 中的原子 token，验证顺序
tokens = re.findall(r'\[.*?\]|Cl|Br|[BCNOSPFIbcnops]', smiles)

# 3. 用 atomNote 属性在图上标注 SMILES 顺序（+1 转成 1-based）
mol_draw = Chem.Mol(mol)
for atom in mol_draw.GetAtoms():
    atom.SetProp('atomNote', str(atom.GetIdx() + 1))  # 1-based SMILES 顺序

# 4. 绘图
drawer = rdMolDraw2D.MolDraw2DCairo(700, 600)
drawer.drawOptions().annotationFontScale = 0.9
rdMolDraw2D.PrepareAndDrawMolecule(drawer, mol_draw)
drawer.WriteDrawingText('mol_smiles_order.png')
```
